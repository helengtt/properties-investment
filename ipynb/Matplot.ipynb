# Plots with Matplotlib and Pyplot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sqlalchemy import create_engine

engine = create_engine('postgresql://postgres:password@this_postgres')
```

```python
df = pd.read_sql('select suburb_id, ttl_house_tier3 as cnt, house_affordability_tier3*100 as aff from affordability where house_affordability_tier3*100 >= 0',engine)
```

```python
df.plot.scatter(x = 'aff', y = 'cnt')
```

```python
properties = pd.read_csv('../data/properties_cleaned.csv')
properties.groupby('property_type')['listing_id'].count().plot(kind='bar')
```
